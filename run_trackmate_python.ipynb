### Load data and segment for input to tracker
Run raw tif files through Cellpose to segment

```python
# dummy path for pretrained model; using Cellpose's existing cyto model instead
segmentation_model = Path('') 
# read in data
for tiff_file in os.listdir(stack_base_path):
    if not tiff_file.endswith(".tif"):
        continue
    print("Segmenting ", tiff_file)
    stack_filepath = Path(os.path.join(stack_base_path, tiff_file))
    stack = utils.read_stack(stack_filepath)
    # run CellPose segmentation
    utils.segment_stack(stack_filepath, segmented_base_path, segmentation_model, diam_px)
    segmented_path = segmented_base_path / f"{stack_filepath.stem}_segmented.tif"
    segmented_stack = utils.read_stack(segmented_path)
    # view overlay of segmentation mask with original image
    utils.view_segmentation_overlay([stack],[segmented_stack],frame=1)
```
